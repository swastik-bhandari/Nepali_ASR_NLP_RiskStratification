```python
"""
predict_from_audio.py
=====================
Full pipeline:
  Audio file (MP3)
    → AssemblyAI transcription  (Nepali speech-to-text)
    → Normalization              (normalization_map.json)
    → Tokenization               (whitespace + em-dash split)
    → Stopword removal           (normalized_stopwords.json)
    → Suffix stripping           (normalized_suffixes.json)
    → SVM prediction             (svm_model.pkl)
    → Output: low / medium / high

Usage
-----
  python predict_from_audio.py                    # uses default AUDIO_FILE below
  python predict_from_audio.py my_recording.mp3  # pass file as argument

Required files (same folder or adjust paths below)
---------------------------------------------------
  svm_model.pkl
  normalization_map.json
  normalized_stopwords.json
  normalized_suffixes.json
"""

import sys
import os
import re
import json
import time
import pickle
import unicodedata
import requests

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# CONFIGURATION — edit these paths to match your setup
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
API_KEY          = "211d63664dfd4079bb1bea18cea0adef"
BASE_URL         = "https://api.assemblyai.com"
HEADERS          = {"authorization": API_KEY}

AUDIO_FILE       ="healthvoice.mp3"

MODEL_PATH       = "svm_model.pkl"
NORM_MAP_PATH    = "normalization_map.json"
STOPWORDS_PATH   = "normalized_stopwords.json"
SUFFIXES_PATH    = "normalized_suffixes.json"

MIN_ROOT_LEN     = 2   # don't strip suffix if root would be shorter than this

# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# STEP 0 — LOAD MODEL + NLP ASSETS
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
print("Loading model and NLP assets...")

model        = pickle.load(open(MODEL_PATH, "rb"))
norm_map     = json.load(open(NORM_MAP_PATH,  encoding="utf-8"))
stopword_set = set(json.load(open(STOPWORDS_PATH, encoding="utf-8"))["stopwords"])
suffix_list  = json.load(open(SUFFIXES_PATH,  encoding="utf-8"))["suffixes"]
# suffix_list is already sorted longest-first (saved that way by nlp_pipeline.py)

_sorted_keys = sorted(norm_map.keys(), key=len, reverse=True)
_protected   = set("".join(norm_map.values()))  # chars used inside replacement values

print("✅  All assets loaded.\n")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# NLP HELPERS  (identical logic to the training pipeline)
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

def normalize(text: str) -> str:
    """Apply normalization_map rules (longest-key first)."""
    text = unicodedata.normalize("NFC", text)
    for key in _sorted_keys:
        rep = norm_map[key]
        if rep == "" and key in _protected:
            continue   # don't strip chars that appear inside replacement values
        if key in text:
            text = text.replace(key, rep)
    return " ".join(text.split())


def tokenize(text: str) -> list:
    """Whitespace tokenizer; also splits on em-dash / hyphen."""
    tokens = []
    for t in text.split():
        parts = re.split(r"[–\-]", t)
        tokens.extend(p for p in parts if p)
    return tokens


def remove_stopwords(tokens: list) -> list:
    return [t for t in tokens if t not in stopword_set]


def strip_suffix(token: str):
    """Return (root, matched_suffix) or (token, None)."""
    for sfx in suffix_list:
        if token.endswith(sfx) and len(token) - len(sfx) >= MIN_ROOT_LEN:
            return token[: -len(sfx)], sfx
    return token, None


def stem_tokens(tokens: list) -> list:
    return [strip_suffix(t)[0] for t in tokens]


def preprocess(raw_text: str) -> str:
    """
    Full NLP pipeline: raw text → single string ready for the SVM.
    This is exactly what was applied to training data.
    """
    normalized = normalize(raw_text)
    tokens     = tokenize(normalized)
    tokens     = remove_stopwords(tokens)
    tokens     = stem_tokens(tokens)
    return " ".join(tokens)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# STEP 1 — UPLOAD AUDIO
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
if not os.path.exists(AUDIO_FILE):
    raise FileNotFoundError(
        f"❌  Audio file '{AUDIO_FILE}' not found.\n"
        f"    Place it in the same folder as this script,\n"
        f"    or pass the path as:  python predict_from_audio.py /path/to/file.mp3"
    )

print(f"⏫  Uploading '{AUDIO_FILE}'...")
with open(AUDIO_FILE, "rb") as f:
    upload_resp = requests.post(BASE_URL + "/v2/upload", headers=HEADERS, data=f)
upload_resp.raise_for_status()
audio_url = upload_resp.json()["upload_url"]
print("✅  Upload complete.")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# STEP 2 — REQUEST TRANSCRIPTION
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
data = {
    "audio_url":          audio_url,
    "speech_models":      ["universal-2"],
    "language_detection": True,
    "language_detection_options": {
        "expected_languages": ["ne"],
        "fallback_language":  "ne",
    },
}
print("📤  Requesting Nepali transcription...")
transcript_resp = requests.post(
    BASE_URL + "/v2/transcript",
    json=data,
    headers={**HEADERS, "content-type": "application/json"},
)
if not transcript_resp.ok:
    print(f"❌  Error {transcript_resp.status_code}: {transcript_resp.text}")
    transcript_resp.raise_for_status()

transcript_id = transcript_resp.json()["id"]
print(f"🆔  Transcript ID: {transcript_id}")


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# STEP 3 — POLL UNTIL DONE
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
polling_url = BASE_URL + "/v2/transcript/" + transcript_id
print("⏳  Transcribing", end="", flush=True)
while True:
    result = requests.get(polling_url, headers=HEADERS).json()
    if result["status"] == "completed":
        print("\n✅  Transcription done.\n")
        raw_text      = result["text"]
        detected_lang = result.get("language_code", "unknown")
        break
    elif result["status"] == "error":
        raise RuntimeError(f"❌  Transcription failed: {result['error']}")
    else:
        print(".", end="", flush=True)
        time.sleep(3)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# STEP 4 — NLP PREPROCESSING
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
normalized_text  = normalize(raw_text)
tokens_raw       = tokenize(normalized_text)
tokens_no_sw     = remove_stopwords(tokens_raw)
tokens_stemmed   = stem_tokens(tokens_no_sw)
model_input      = " ".join(tokens_stemmed)


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# STEP 5 — PREDICT
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
prediction = model.predict([model_input])[0]

SEVERITY_EMOJI = {"low": "🟢", "medium": "🟡", "high": "🔴"}
SEVERITY_DESC  = {
    "low":    "Routine — no immediate action needed",
    "medium": "Moderate — monitor closely, consult if worsens",
    "high":   "Urgent — immediate medical attention required",
}


# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
# STEP 6 — OUTPUT
# ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
SEP = "=" * 60
print(SEP)
print("🌐  Detected language   :", detected_lang)
print()
print("📝  Raw transcription   :")
print("   ", raw_text)
print()
print("🔤  Normalized          :")
print("   ", normalized_text)
print()
print("🪙  Tokens (stemmed)    :")
print("   ", tokens_stemmed)
print()
print(SEP)
print(f"🏥  SEVERITY PREDICTION :  {SEVERITY_EMOJI[prediction]}  {prediction.upper()}")
print(f"    {SEVERITY_DESC[prediction]}")
print(SEP)

# Save full result to JSON
output = {
    "audio_file":       AUDIO_FILE,
    "detected_language": detected_lang,
    "raw_transcription": raw_text,
    "normalized_text":  normalized_text,
    "tokens_raw":       tokens_raw,
    "tokens_no_sw":     tokens_no_sw,
    "tokens_stemmed":   tokens_stemmed,
    "model_input":      model_input,
    "prediction":       prediction,
    "description":      SEVERITY_DESC[prediction],
}
out_file = "prediction_result.json"
with open(out_file, "w", encoding="utf-8") as f:
    json.dump(output, f, ensure_ascii=False, indent=2)
print(f"\n💾  Full result saved to '{out_file}'")
```
