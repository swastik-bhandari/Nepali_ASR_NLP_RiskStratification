```python
import csv
import json
import re
from collections import Counter
```

```python
class NepaliTextPreprocessor:
    """
    A comprehensive preprocessor for Nepali text that performs:
    1. Normalization (character-level replacements)
    2. Tokenization (word segmentation)
    3. Stop word removal
    """
    
    def __init__(self, normalization_file, suffixes_file, stopwords_file):
        """Load normalization map, suffixes, and stopwords from JSON files"""
        
        # Load normalization map
        with open(normalization_file, 'r', encoding='utf-8') as f:
            self.normalization_map = json.load(f)
        
        # Load suffixes
        with open(suffixes_file, 'r', encoding='utf-8') as f:
            data = json.load(f)
            self.suffixes = set(data['suffix'])
        
        # Load stopwords
        with open(stopwords_file, 'r', encoding='utf-8') as f:
            data = json.load(f)
            self.stopwords = set(data['stopwords'])
        
        print("✓ Loaded normalization map with {} mappings".format(len(self.normalization_map)))
        print("✓ Loaded {} suffixes".format(len(self.suffixes)))
        print("✓ Loaded {} stopwords".format(len(self.stopwords)))
        print()
    
```

```python
    def normalize(self, text):
        """
        Apply character-level normalization to the text.
        Multi-character sequences are replaced first, then single characters.
        """
        # Sort keys by length (longest first) to handle multi-character sequences
        sorted_keys = sorted(self.normalization_map.keys(), key=len, reverse=True)
        
        normalized_text = text
        for char in sorted_keys:
            normalized_text = normalized_text.replace(char, self.normalization_map[char])
        
        # Remove extra whitespace
        normalized_text = ' '.join(normalized_text.split())
        
        return normalized_text
```

```python
    def tokenize(self, text):
        """
        Tokenize Nepali text into words.
        Splits on whitespace and removes empty tokens.
        """
        # Split on whitespace
        tokens = text.split()
        
        # Remove empty tokens
        tokens = [token for token in tokens if token.strip()]
        
        return tokens
```

```python
    def remove_stopwords(self, tokens):
        """
        Remove stopwords from the token list.
        """
        filtered_tokens = [token for token in tokens if token not in self.stopwords]
        return filtered_tokens
```

```python
    def preprocess(self, text):
        """
        Complete preprocessing pipeline:
        1. Normalize
        2. Tokenize
        3. Remove stopwords
        
        Returns a dictionary with all intermediate results for debugging.
        """
        # Step 1: Normalize
        normalized = self.normalize(text)
        
        # Step 2: Tokenize
        tokens = self.tokenize(normalized)
        
        # Step 3: Remove stopwords
        filtered_tokens = self.remove_stopwords(tokens)
        
        return {
            'original': text,
            'normalized': normalized,
            'tokens': tokens,
            'filtered_tokens': filtered_tokens,
            'token_count': len(tokens),
            'filtered_count': len(filtered_tokens)
        }
```

```python
def process_csv_dataset(input_file, output_file, preprocessor):
    """
    Process the CSV dataset and create a new CSV with preprocessed text.
    """
    
    processed_data = []
    
    with open(input_file, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        rows = list(reader)
    
    print(f"Processing {len(rows)} documents from {input_file}...")
    print("=" * 80)
    print()
    
    # Process each row
    for i, row in enumerate(rows, 1):
        text = row['text']
        label = row['label']
        
        # Preprocess the text
        result = preprocessor.preprocess(text)
        
        # Store processed data
        processed_data.append({
            'original_text': result['original'],
            'normalized_text': result['normalized'],
            'tokens': ' '.join(result['tokens']),
            'filtered_tokens': ' '.join(result['filtered_tokens']),
            'token_count': result['token_count'],
            'filtered_count': result['filtered_count'],
            'label': label
        })
        
        # Print first 5 examples for verification
        if i <= 5:
            print(f"Example {i} (Label: {label}):")
            print(f"  Original:  {result['original'][:80]}...")
            print(f"  Normalized: {result['normalized'][:80]}...")
            print(f"  Tokens ({result['token_count']}): {' '.join(result['tokens'][:10])}...")
            print(f"  Filtered ({result['filtered_count']}): {' '.join(result['filtered_tokens'][:10])}...")
            print()
    
    # Write to output CSV
    with open(output_file, 'w', newline='', encoding='utf-8') as f:
        fieldnames = ['original_text', 'normalized_text', 'tokens', 'filtered_tokens', 
                      'token_count', 'filtered_count', 'label']
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(processed_data)
    
    print("=" * 80)
    print(f"✓ Processed {len(processed_data)} documents")
    print(f"✓ Output saved to: {output_file}")
    print()
    
    # Statistics
    total_tokens_before = sum(row['token_count'] for row in processed_data)
    total_tokens_after = sum(row['filtered_count'] for row in processed_data)
    
    print("Statistics:")
    print(f"  Total tokens before stopword removal: {total_tokens_before}")
    print(f"  Total tokens after stopword removal: {total_tokens_after}")
    print(f"  Tokens removed: {total_tokens_before - total_tokens_after}")
    print(f"  Reduction: {((total_tokens_before - total_tokens_after) / total_tokens_before * 100):.2f}%")
    print()
    
    # Label distribution
    label_counts = Counter(row['label'] for row in processed_data)
    print("Label distribution:")
    for label, count in sorted(label_counts.items()):
        print(f"  {label}: {count}")
    
    return processed_data
```

```python
def NepaliPreprocessing():
    """Main execution function"""
    
    print("=" * 80)
    print("NEPALI TEXT PREPROCESSING PIPELINE")
    print("=" * 80)
    print()
    
    # Initialize preprocessor
    preprocessor = NepaliTextPreprocessor(
        normalization_file='normalization_map.json',
        suffixes_file='suffixes.json',
        stopwords_file='stopwords.json'
    )
    
    # Process the dataset
    input_csv = 'dataset.csv'
    output_csv = 'dataset_preprocessed.csv'
    
    processed_data = process_csv_dataset(input_csv, output_csv, preprocessor)
    
    # Test with the example from the document
    print()
    print("=" * 80)
    print("VERIFICATION WITH PROVIDED EXAMPLE:")
    print("=" * 80)
    test_text = "महिलालाई छाती दुखिरहेको छ र सास फेर्न गाह्रो छ।"
    expected = "महिलालाइ छाति दुखिरहेको छ र सास फेर्न गाह्रो छ"
    
    result = preprocessor.preprocess(test_text)
    
    print(f"Original:  {result['original']}")
    print(f"Normalized: {result['normalized']}")
    print(f"Expected:   {expected}")
    print(f"Match: {'✓ YES' if result['normalized'] == expected else '✗ NO'}")
    print()
    print(f"Tokens: {result['tokens']}")
    print(f"Filtered tokens: {result['filtered_tokens']}")
    print()
    print("=" * 80)
```

```python
NepaliPreprocessing()
```
